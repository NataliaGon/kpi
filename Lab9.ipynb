```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report


(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Нормалізація пікселів до [0,1]
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Перетворюємо мітки у one-hot формат (10 класів)
y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)
```

# Побудова нейронної мережі (MLP)

```python
model = Sequential([
    Flatten(input_shape=(28, 28)),     # 28x28 -> 784
    Dense(128, activation="relu"),     # прихований шар
    Dense(64, activation="relu"),      # ще один прихований шар (можна забрати)
    Dense(10, activation="softmax")    # вихідний шар, 10 класів
])

```

# Компіляція моделі, функція втрат + оптимізатор + метрики

```python
model.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

# Навчання моделі

```python
history = model.fit(
    x_train, y_train_cat,
    batch_size=128,
    epochs=10,
    validation_split=0.1,
    verbose=1
)

```

# Оцінка моделі на тестових даних

```python
test_loss, test_acc = model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Test loss: {test_loss:.4f}")
print(f"Test accuracy: {test_acc:.4f}")

y_pred_probs = model.predict(x_test)
y_pred = np.argmax(y_pred_probs, axis=1)

print(classification_report(y_test, y_pred, digits=4))

```

Отримані результати свідчать про високу якість моделі: точність на тестовому наборі перевищує 97 %, а значення precision, recall та F1-показника залишаються стабільно високими для всіх класів. Мережа продемонструвала здатність коректно узагальнювати інформацію та ефективно розпізнавати нові зображення, що підтверджує адекватність обраної архітектури, правильність підготовки даних та ефективність навчання. Отже, створена модель успішно виконує поставлене завдання та може бути використана як базове рішення для задач класифікації зображень рукописних цифр.
